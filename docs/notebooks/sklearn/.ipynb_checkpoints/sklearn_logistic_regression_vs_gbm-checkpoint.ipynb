# Example using scikit-learn: Logistic Regression model vs Gradient Boosting model, Pre-processing feature correlations heatmap, GridSearch models, Feature importances, Percentile bins

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import GridSearchCV

from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import roc_curve, auc

from sklearn.preprocessing import label_binarize, LabelBinarizer
```

## Get the dataset and describe it

```python
dat = datasets.load_breast_cancer()
# print(dat.DESCR)
```

## Use pandas to describe columns

```python
print("dataset keys:")
print(dat.keys()) # dict_keys(['target_names', 'target', 'feature_names', 'data', 'DESCR'])

li_classes = list(dat.target_names)
li_target = list(dat.target)
li_ftrs = list(dat.feature_names)

print("li_classes", li_classes)
print("li_target[:30]", li_target[:30])

df_all = pd.DataFrame(dat.data[:,:], columns=li_ftrs)
print()
print("describe dataframe, first five columns:")
print(df_all.iloc[:,:5].describe())
```

## Check feature correlations

```python
# Setup X and y
X = df_all
y = pd.Series(li_target)

# Don't look for correlations between dummy features
cols_dummies = []
cols_corr = [c for c in X.columns if c not in cols_dummies]
Xcorr = X[cols_corr].corr()

print(len(X.columns))
print(cols_dummies)
print(len(cols_dummies))
print(cols_corr)
print(len(cols_corr))

dfcorr = np.abs(Xcorr)
plt.figure(figsize=(10, 10))
plt.imshow(dfcorr, cmap='RdYlGn', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(dfcorr)), dfcorr.columns, rotation='vertical')
plt.yticks(range(len(dfcorr)), dfcorr.columns);
plt.suptitle('Features Correlations Heat Map', fontsize=15, fontweight='bold')
plt.show()
```

## Highest correlated feature pairs

```python
CORRELATION_MAX = 0.8
s_pairs = dfcorr.unstack()
s_pairs = np.round(s_pairs, decimals=3)
s_pairs_sorted = s_pairs.order(kind="quicksort", ascending=False)
s_pairs_sorted = s_pairs_sorted[(s_pairs_sorted != 1) & (s_pairs_sorted > CORRELATION_MAX)]
# print(s_pairs_sorted)
li_corr_pairs = s_pairs_sorted.index.tolist()
print(s_pairs_sorted)
```

## Features most correlated with the target

```python
s_corr_target = Xcorr.tail(1)
s_corr_target = np.round(s_corr_target, decimals=3)

c = []
for cols in s_corr_target:
    if s_corr_target[cols][0] >= 0.1:
        c.append(cols)
     
s_corr_target_sort = s_corr_target.iloc[0,:].sort_values()
print(s_corr_target_sort.head(20))
print(s_corr_target_sort.tail(20))
```

## Absolute sorted values

```python
s_corr_target_abs = np.abs(s_corr_target_sort).sort_values()
print(s_corr_target_abs[::-1])
```

## For the highest correlated pairs, remove the feature that is less correlated with the target variable

```python
li_remove = []
li_remove_scores = []
for tup in li_corr_pairs:
    s0 = s_corr_target_abs.loc[tup[0]]
    s1 = s_corr_target_abs.loc[tup[1]]
    rem = tup[1] if s1 < s0 else tup[0]  # add the feature that is less correlated with the target variable
    if rem not in li_remove:
        li_remove.append(rem)
        li_remove_scores.append((tup[0], tup[1], s0, s1))
   
print("li_remove")
print(li_remove)
for v in li_remove_scores:
    print(v)

print()
li_final_cols = list(set(X) - set(li_remove)) 
li_final_cols.sort()
print("li_final_cols")
print(li_final_cols)

print()
X = X[li_final_cols]
print("After removing correlated features, X.shape:", X.shape)
```

## Print target variable stats

```python
val_cnts = y.value_counts()
print(val_cnts)
print("Positive percentage of total applicants:")
print("%2f%%" % (100 * val_cnts[1] / len(y)))
```

## Split train and test sets

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
print(y_test.tail(10))
# print(X_test.tail(10))
```

## Build Logistic Regression Classifier

```python
grid_lr = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}
gs_lr = GridSearchCV(LogisticRegression(class_weight='balanced', dual=False, 
          fit_intercept=True, intercept_scaling=1, max_iter=200,
          n_jobs=1, penalty='l2', random_state=0,
          tol=0.0001, verbose=0, warm_start=False),
          grid_lr, verbose=1)
gs_lr.fit(X_train, y_train)

print("gs_lr.best_estimator_", gs_lr.best_score_)
clf_lr = gs_lr.best_estimator_
print("gs_lr.best_estimator_")
print(clf_lr)
print(clf_lr.intercept_ )
print(clf_lr.coef_)
```

## Build GB Classifier

```python
y_true_cnt = y_train.value_counts()[1]
y_len = len(y_train)
y_true_ratio = y_true_cnt / y_len
y_true_ratio_comp = 1 - y_true_ratio

print("y_true_cnt, y_len, y_true_ratio, y_true_ratio_comp")
print(y_true_cnt, y_len, y_true_ratio, y_true_ratio_comp)

TRUE_CLASS_WEIGHT = y_true_ratio_comp
FALSE_CLASS_WEIGHT = 1 - TRUE_CLASS_WEIGHT
sample_weight = [TRUE_CLASS_WEIGHT if y == 1 else FALSE_CLASS_WEIGHT for y in y_train]

grid_gb = {'min_samples_leaf': [2, 4, 8, 16]}

gs_gb = GridSearchCV(GradientBoostingClassifier(criterion='friedman_mse',
              init=None, loss='deviance', max_features='sqrt', learning_rate=0.2,
              max_leaf_nodes=None, max_depth=3,
              min_impurity_decrease=0.0, min_impurity_split=None,
              min_samples_split=2,
              min_weight_fraction_leaf=0.0, n_estimators=200,
              presort='auto', random_state=0, subsample=1.0, verbose=0,
              warm_start=False),
              grid_gb, verbose=0)

gs_gb.fit(X_train, y_train, sample_weight)

print("gs_gb.best_score_", gs_gb.best_score_)
clf_gb = gs_gb.best_estimator_
print("gs_gb.best_estimator_")
print(clf_gb)
```

## Feature importances

```python
# https://stackoverflow.com/questions/34052115/how-to-find-the-importance-of-the-features-for-a-logistic-regression-model
dfimp = pd.DataFrame(np.std(X_train, 0), columns=['std'])
dfimp['coef'] = clf_lr.coef_.T
dfimp['lr_sign_imp'] = (np.std(X_train, 0).ravel() * clf_lr.coef_).T
dfimp['lr_imp'] = np.abs(dfimp['lr_sign_imp'])
dfimp['gb_imp'] = pd.Series(clf_gb.feature_importances_, X_train.columns).sort_values(ascending=False)

dfimp['lr_rank'] = dfimp['lr_imp'].rank()
dfimp['gb_rank'] = dfimp['gb_imp'].rank()

dfsort = dfimp.sort_values('gb_rank', ascending=False)
print(dfsort.iloc[:20,:].to_string())
```

## First rows in X and y 

```python
cols_top = dfsort.index[:7]
print("X_test")
print(X_test[cols_top].iloc[-10:,:].to_string())
print()
print("y_test")
print(y_test.iloc[-10:])
```

## Predict scores, print Confusion Matrix

```python
li_clfs = ['clf_lr', 'clf_gb'] #, 'clf_dt']
dfp = pd.DataFrame(index=['TARGET'], data=[y_test]).T

for s_clf in li_clfs:
    print("MODEL: " + s_clf)
    print("--------------")
    clf = eval(s_clf)
    
    y_pred = clf.predict(X_test).astype(int)
    y_score = clf.predict_proba(X_test)

    s_class = "%s_class" % s_clf
    s_proba = "%s_proba" % s_clf
    s_rank = "%s_rank" % s_clf
    
    dfp[s_class] = y_pred
    dfp[s_proba] = y_score[:,1]
    dfp[s_rank] = dfp[s_proba].rank(ascending=1).astype(int)
    
    # Print confusion matrix & classification report
    cm = confusion_matrix(y_test, y_pred)
    print("Confusion table: (ROWS 0,1: reality, COLS 0,1: prediction)")
    print(cm) 
    print("Classification report:")
    print(classification_report(y_test, y_pred))
    
    # Binarize the output
    lb = LabelBinarizer()
    y_test_bin = lb.fit_transform(y_test)
    y_test_bin = np.hstack((1-y_test_bin, y_test_bin))
#     y_test_bin = label_binarize(y_test, classes=[0, 1, 2])
    
    if s_clf == 'clf_lr':
        y_score_lr = y_score.copy() 
    elif s_clf == 'clf_gb':
        y_score_gb = y_score.copy()
    else:
        print('Error')
        break
```

## Percentage of positives in test set

```python
cm0_support = cm[0,0] + cm[0,1]
cm1_support = cm[1,0] + cm[1,1]
cm_total = cm0_support + cm1_support
perc_pos = 100 * cm1_support / cm_total
print("cm0_support, cm1_support, cm_total: ", cm0_support, cm1_support, cm_total)
print("Percentage of positives: ", perc_pos)
```

## Print important feature for a few records

```python
print(dfp.tail(10).to_string())   
```

## Percentile bins

```python
RANK_BINS = 5
rank_cols = ['i_bin', 'i_min_bin', 'i_max_bin', 'score_max', 'bin_cnt', 'Pos_cnt', 'Pos_rate']
rank_cols2 = ['i_bin', 'score_max', 'tnr', 'fpr', 'fnr', 'tpr', 'tn', 'fp', 'fn', 'tp'] 

print(dfp.shape)
len_test = dfp.shape[0]
len_bin = int(len_test / RANK_BINS)

for s_clf in li_clfs:
    i_min_bin = 0
    i_max_bin = 0
    dfr = pd.DataFrame(columns=rank_cols)
    dfr2 = pd.DataFrame(columns=rank_cols2)
    
#     print("MODEL: " + s_clf)
#     print("--------------")
    
    s_class = "%s_class" % s_clf
    s_proba = "%s_proba" % s_clf
    s_rank = "%s_rank" % s_clf
    
    for i in range(RANK_BINS):
        if i == RANK_BINS - 1:
            i_max_bin = len_test
        else:
            i_max_bin += len_bin

        # Range used for *each* risk bin
        df_rng = dfp[(dfp[s_rank] >= i_min_bin) & (dfp[s_rank] < i_max_bin)]
        
        score_max = np.max(df_rng[s_proba])
        bin_cnt = df_rng.shape[0]
            
        Pos_cnt = len(df_rng[df_rng['TARGET'] == 1])
        Pos_rate = Pos_cnt / bin_cnt
        
        # Range used for *all* risk bins up to i_max_bin
        df_rng_0 = dfp[dfp[s_rank] < i_max_bin]
        df_rng_1 = dfp[dfp[s_rank] >= i_max_bin]
        
#         pred_0 = len(df_rng_0)
#         pred_1 = len(df_rng_1)
        
        # Positive targets (positive customers)
        tp = len(df_rng_1[df_rng_1['TARGET'] == 1])
        fn = len(df_rng_0[df_rng_0['TARGET'] == 1])
        pos = tp + fn
        tpr = tp / pos
        fnr = 1 - tpr
        
        # Negative targets
        tn = len(df_rng_0[df_rng_0['TARGET'] == 0])
        fp = len(df_rng_1[df_rng_1['TARGET'] == 0])
        neg = tn + fp
        fpr = fp / neg      
        tnr = 1 - fpr
        
        row = [i, i_min_bin, i_max_bin, score_max, bin_cnt, Pos_cnt, Pos_rate]
        dfr.loc[i] = row
        
        row2 = [i, score_max, tnr, fpr, fnr, tpr, tn, fp, fn, tp]
        dfr2.loc[i] = row2
        
        # Prep for next iteration
        i_min_bin = i_max_bin 
              
    if s_clf == 'clf_lr':
        dfr_lr = dfr.copy() 
        dfr2_lr = dfr2.copy() 
    elif s_clf == 'clf_gb':
        dfr_gb = dfr.copy()
        dfr2_gb = dfr2.copy()
    else:
        print('Error')
        break
 
print("-------")
print("Results of positive stats per bin")
print("Logistic Regression stats:")
print(dfr_lr.to_string())
print()
print("GBM stats:")
print(dfr_gb.to_string())

print()
print("-------")
print("Results of cumulutive positive stats")
print("Logistic Regression stats:")
print(dfr2_lr.to_string())
print()
print("GBM stats:")
print(dfr2_gb.to_string())

print()
print("-------")
print("Overall CO stats in the test set:")
Pos_cnt = len(dfp[dfp['TARGET'] == 1])
Pos_rate = Pos_cnt / len_test

print("Pos_cnt", Pos_cnt)
print("Total rows", len_test)
print("Pos_rate", Pos_rate)
```

## Plot each model's chart by percentiles

```python
s_titles = ['Logistic Regression', 'GBM']

plt.figure(figsize=(12, 4))
for i, s_clf in enumerate(li_clfs):
    s_title = s_titles[i]
    
    if s_clf == 'clf_lr':
        dfr = dfr_lr 
        dfr2 = dfr2_lr 
    elif s_clf == 'clf_gb':
        dfr = dfr_gb
        dfr2 = dfr2_gb
    else:
        print('Error')
        break
    
    plt.subplot(1, 2, i+1)
    plt.bar(dfr['i_bin'], dfr['Pos_rate'])
    plt.title('Percentiles for ' + s_title)

    # Format charts
    plt.grid()
    plt.xlabel('Risk Bin (from low to high scoring bins)')
    plt.ylabel('Ratio of positives per bin')
    plt.legend(loc="lower right")
    
    plt.xlim([0, 5])
    plt.ylim([0.0, 1])
    
plt.show()
```
