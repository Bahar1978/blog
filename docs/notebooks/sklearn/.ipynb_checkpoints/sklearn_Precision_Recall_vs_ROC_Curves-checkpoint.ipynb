# Example using scikit-learn: Precision Recall vs ROC Curves

```python
import os
import matplotlib.pyplot as plt
import pandas as pd
import random
import numpy as np
np.set_printoptions(precision=3)

from scipy.stats import norm
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
from sklearn.metrics import precision_recall_curve
from sklearn.metrics import roc_curve, auc
```

## Generate normally distributed data

```python
BINS = 20

# Distribution #1
N1 = 10000
mu1 = -1
sigma1 = 1
s1 = np.sort(norm.rvs(loc=mu1, scale=sigma1, size=N1))
plt.hist(s1, normed=0, color='green', alpha=.5, bins=BINS)

x1 = np.linspace(-4, 4, N1)
bell1 = norm.pdf(x1) * N1 * max(norm.pdf(x1))
plt.plot(x1-1, bell1, label='S1', color='green')

# Distribution #2
N2 = 10000
mu2 = 0
sigma2 = 1
s2 = np.sort(norm.rvs(loc=mu2, scale=sigma2, size=N2))
plt.hist(s2, normed=0, color='orange', alpha=.5, bins=BINS)

x2 = np.linspace(-4, 4, N2)
bell2 = norm.pdf(x2) * N2 * max(norm.pdf(x2))
plt.plot(x2, bell2, label='S2', color='orange')

# Show
plt.title("Normal distribution plots")
# plt.grid()
plt.legend()
plt.show()

# Stats
# print(pd.Series(s1).describe())
# print(pd.Series(s2).describe())
```

## ROC and Precision Recall Curves

```python
y_true = N1*[0] + N2*[1] 
y_score = list(s1) + list(s2)

fpr, tpr, _ = roc_curve(y_true, y_score)
roc_auc = auc(fpr, tpr)

# plt.figure(figsize=(10, 8))

plt.subplot(1,2,1)
lw = 2
plt.plot(fpr, tpr, color='darkorange', lw=lw, label='ROC curve (area = %0.4f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc="lower right")
plt.grid()

plt.subplot(1,2,2)
precision, recall, _ = precision_recall_curve(y_true, y_score)
# print("precision:", precision)
# print("recall:", recall)

plt.step(recall, precision, color='orange', where='post')
# plt.fill_between(recall, precision, step='post', alpha=0.5, color='orange')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.ylim([0.0, 1.05])
plt.xlim([0.0, 1.0])
plt.title('Precision-Recall Curve')
plt.grid()

left  = 0.125  # the left side of the subplots of the figure
right = 0.9    # the right side of the subplots of the figure
bottom = 0.1   # the bottom of the subplots of the figure
top = 0.9      # the top of the subplots of the figure
wspace = 0.5   # the amount of width reserved for blank space between subplots
hspace = 0.2   # the amount of height reserved for white space between subplots
plt.subplots_adjust(left, bottom, right, top, wspace, hspace)
plt.show()
```
