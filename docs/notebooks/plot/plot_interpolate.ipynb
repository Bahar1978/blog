# Example plots using scipy interpolate

```python
# Setup
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Build data
XMIN = 0
XMAX = 11
STEP = 3

x = np.arange(XMIN, XMAX, STEP)
y = np.exp(-x/1.2)  # Original y data is for an exponential function

print("x:", x)
print("y:", y)
```

## Interpolate data using different kinds of methods

```python
# Plot the original data
fig = plt.figure()
ax = plt.gca()
plt.plot(x, y, 's', label='original')
plt.grid(True)

# Interpolate
for k in ['linear', 'quadratic', 'cubic']:
    f = interpolate.interp1d(x, y, kind=k) 
    
    STEP_NEW = 1
    x_new = np.arange(XMIN, XMAX - STEP_NEW, STEP_NEW)
    y_new = f(x_new)  
    plt.plot(x_new, y_new, '-o', label=k)
    
print('x', x)
print('x_new', x_new)

plt.suptitle("Curve Fitting Example", color='m')
plt.legend(loc='upper right')
ax.set_xlabel('x', color='m')
ax.set_ylabel('y', color='m')
plt.show()
```
