# Example plots using scipy.interpolate

```python
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

# Build data
XMIN = 0
XMAX = 11
STEP = 3
STEP_NEW = 1

x = np.arange(XMIN, XMAX, STEP)
y = np.exp(-x/1.2)

# Plot the original data
fig = plt.figure()
ax = plt.gca()
plt.plot(x, y, 'o', label='orig')
plt.grid()

# Interpolate date for different 'kind' parameters
for k in ['linear', 'quadratic', 'cubic']:
    f = interpolate.interp1d(x, y, kind=k) 
    xnew = np.arange(XMIN, XMAX - STEP_NEW, STEP_NEW)
    ynew = f(xnew)  
    plt.plot(xnew, ynew, '-x', label=k)
    
print('x', x)
print('xnew', xnew)

plt.suptitle("Curve Fitting Example", color='m')
plt.legend(loc='upper right')
ax.set_xlabel('x1')
ax.set_ylabel('y1')
plt.show()
```
