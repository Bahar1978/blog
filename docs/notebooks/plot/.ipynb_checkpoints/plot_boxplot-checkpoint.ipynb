# Example of pandas boxplot

```python
import pandas as pd
import matplotlib.pyplot as plt
import random

plt.figure()

MAX = 500
STEP = 100

li_x = [x for x in range(0, MAX)]
li_y = [round(x + MAX*random.random(),2) for x in range(0, MAX)]
di = {'X': li_x, 'Y': li_y}
df = pd.DataFrame.from_dict(di)
print(df.describe())
```

## Build dataframes

```python
di_x = {}
di_y = {}
for v in range(0, MAX, STEP):
    rng = (df['X'] >= v) & (df['X'] < v + STEP)
    
    sx = df['X'][rng]
    kx = "%d-%d[n=%d]" % (v, v+len(sx)-1, len(sx))
    di_x[kx] = sx

    sy = df['Y'][rng]
    ky = "%d-%d[n=%d]" % (v, v+len(sy)-1, len(sy))
    di_y[ky] = sy

dfx = pd.DataFrame(di_x)
dfy = pd.DataFrame(di_y)

print(dfy.head().to_string())
print()
print(dfy.tail().to_string())
```

## Scatter plot

```python
fig = plt.figure()
plt.grid()
for i, dy in enumerate(dfy):
    x = dfx.iloc[:, i]
    y = dfy[dy]
    plt.plot(x, y, 'bx', alpha=0.9)

plt.show()
```

## Box plot

```python
fig = plt.figure()
ax = plt.gca()

print(dfy.shape)
dfy.boxplot(showmeans=True)
plt.suptitle("Boxplot Example", color='m')
ax.set_xlabel('x_label')
ax.set_ylabel('y_label')

# Rotate x axis text values
for tick in ax.get_xticklabels():
    tick.set_rotation(10)

plt.show()
```
