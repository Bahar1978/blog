# List comprehensions

## Transform 1 dimensional list

```
x1 = [1,2,3,4,5]

x2 = [10*x for x in x1]
x2
```

## Filter out values with 'if'

```
x3 = [x for x in x2 if x>10 and x<50]
x3
```

## Transform values with 'if' and 'else'

```
x4 = [0 if x < 30 else 1 for x in x3]
x4
```

##Create and fill a list of lists

```
x5 = [[] for x in range(5)]
x5
```

```
x5[-1].append(3)
x5
```

```
x5.extend([[6,7]])
x5
```

## Flatten a 2-dimensional array

```
dat = [[1,2,3],[4,5,6]]
x6 = [val for row in dat for val in row]
x6
```

## Find all the indexes where list values changed in sequential order

```
aa = [1,1,2,3,3,3,4]
x7 = [0] + [i+1 for i,x in enumerate(aa[1:]) if x != aa[i]]
x7
```

```
# Equivalent to:
x8 = [0] + [i+1 for i,x in enumerate(aa[:-1]) if x != aa[i+1]]
x8
```

## Get value pairs from two lists using 'zip'

```
xx = [11,12,13]
yy = [21,22,23]
xy1 = list(zip(xx, yy))
xy1
```

## Iterate 2 lists to get all possible pair tuples

```
# Equivalent to using 2 for loops as follows:
# for x in xx:
#     for y in yy:
#        xy = (x,y)

xy2 = [(x,y) for x in xx for y in yy]
xy2
```
