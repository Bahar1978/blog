# Example of how to setup a class framework for use in notebooks, part 1 of 2

## Create a directory for python modules

```
mkdir -p tmp/modules  # -p option will create parent directory first, if it doesn't exist
```

## Write a test class to a new file, test_class.py

```
%%writefile tmp/modules/test_class.py

class My_Test_Class:
  
  class_var = 'class variable'
  
  # __init__
  def __init__(self, instance_var):
    self.instance_var = instance_var
    self.instance_list = []
    print("self.instance_var:", self.instance_var)
    
  # Example method
  def add_to_list(self, item_to_add):
    self.instance_list.append(item_to_add)
    print("self.instance_list:", self.instance_list)
    
```

## Writing this file will mark the directory on disk as a python package directory

```
%%writefile tmp/modules/__init__.py
def my_func():
  print("my_func")
```

## List the newly created files

```
ls -la tmp/modules
```
