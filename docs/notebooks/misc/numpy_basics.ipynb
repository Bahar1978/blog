# Numpy basics

```
# Setup
import numpy as np
```

## Create arrays

```
li = [7,8,9]  # python list
ar = np.array(li)  # create a numpy array from a python list
li2 = ar.tolist()  # convert a python list from an array

print("1. li:", li, type(li))
print("2. ar:", ar, type(ar))
print("3. li2:", li2, type(li2))
```

```
a0 = np.zeros((3,2))  # default is a float array of rows by columns
a0
```

```
print("Array info:")
a0.dtype, len(a0), a0.ndim, a0.shape, a0.size
```

```
a1 = np.ones((2,3), dtype=np.int32)  # create as ints
a1
```

```
a2 = np.array([[1,2,3],[4,5,6]])  # 2 dim array
a2
```

```
a3 = np.arange(5,55,10)  # start=5, end=55, step=10  (note that the last element is less than the end value)
a3
```

```
a4 = np.linspace(0.5,5,10)  # 10 equally spaced values from 0.5 to 5 (inclusive)
a4
```

## Access array elements

```
# Setup
a = np.array([[1,2,3],[4,5,6]])
a
```

```
a[-1,-1]  # select last element in the array
```

```
a[-1]  # select last row
```

```
a[1,2]  # select element at [row,col] (note index starts at 0)
```

```
a[:,:2]  # select all rows for the first 2 columns
```

```
a[:,-2:]  # select all rows for the last 2 columns
```

## Boolean indexing of arrays

```
b = a.ravel()
b
```

```
# Boolean index to get array of values
b[(b > 2) & (b < 6)]
```

```
# Boolean index to get array of indexes
np.where((b > 2) & (b < 6))[0]  # note the [0] at the end
```

```
# Element-wise comparison
b2 = np.copy(b)
b2[0] = -999
b == b2
```

## Transform arrays

```
# Convert array to a different type
a.astype(float)
```

```
# Transpose - note that this will cast the float array to integers
b = a.T
b
```

```
# Reshape - note reshape is not equivalent to transpose in the case above
c = a.reshape(3,2)
c
```

## Combine arrays

```
# hstack extends horizontally (maintains row dimension)
np.hstack((b,c))
```

```
np.concatenate((b,c), axis=1) # same as above for axis=1
```

```
# vstack extends vertically (maintains column dimension)
np.vstack((b,c))
```

```
np.concatenate((b,c), axis=0) # same as above for axis=0
```

## More useful functions

```
a = np.array([True, True, False, True])

# Return True if any element of a is True
np.any(a)
```

```
# Return True if all elements of a are True
np.all(a)
```

```
a = np.array([[4,7,9], [5,3,2]])  
np.min(a, axis=None)  # global min value
```

```
np.min(a, axis=0)  # min values along axis=0
```

```
np.min(a, axis=1)  # min values along axis=1
```

```
np.argmin(a, axis=None)  # global min value's global index
```

```
np.argmin(a, axis=0)  # column indexes of min values along axis=0
```

```
np.argmin(a, axis=1)  # row indexes of min values along axis=1
```

```
b = np.array([2.345, 3.456])
np.round(b, 2)
```

## Summary statistics

```
NMAX = 10
N = 30
dat = NMAX * np.random.random(N)  # create a random array of 30 values between 0 and 10
print(dat)
print("---")

from scipy import stats
print(stats.describe(dat))
```

```
# Numpy histogram
np.histogram(dat, bins=np.arange(NMAX))
```

```
# Matplotlib histogram
import matplotlib.pyplot as plt
plt.hist(dat, bins=np.arange(NMAX))  # note that often we add option: normed=True
```
